# 💜 Build Sadie APK
## One-click Android build for your AI proxy agent

**Instructions:**
1. Click **Runtime → Run all** (or press Ctrl+F9)
2. Wait ~15-20 minutes for the first build (it downloads the Android SDK/NDK)
3. Download your APK from the link at the bottom!

This notebook pulls Sadie's source straight from GitHub, so there's nothing to upload.

---


### Step 1: Install Build Tools


```
# Install buildozer + Cython and the native libraries the build needs.
# Package names below work on current Colab (Ubuntu 22.04).
!pip install --quiet buildozer cython
!sudo apt-get update -qq
!sudo apt-get install -y -qq \
    build-essential git zip unzip openjdk-17-jdk \
    autoconf automake libtool pkg-config cmake \
    zlib1g-dev libncurses-dev libtinfo6 libffi-dev libssl-dev lld

# Make sure the build uses Java 17
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
print('✅ Build tools installed! JAVA_HOME =', os.environ['JAVA_HOME'])
```

### Step 2: Get Sadie's Source (from GitHub)


```
# Clone the repo so the build always has the latest source — no manual upload needed.
import os, shutil

REPO_URL = 'https://github.com/stacyfoster30-design/sadie.git'
PROJECT_DIR = '/content/sadie-app'

if os.path.exists(PROJECT_DIR):
    shutil.rmtree(PROJECT_DIR)

!git clone --depth 1 {REPO_URL} {PROJECT_DIR}
os.chdir(PROJECT_DIR)
print('\n📁 Project folder ready:')
!ls -la {PROJECT_DIR}
```

> **Optional:** If you'd rather build from local files instead of GitHub, replace the
> cell above with a `google.colab.files.upload()` step and unzip your
> `sadie-agent-setup.zip` into `/content/sadie-app`.


### Step 3: Build the APK (~15 min first time)


```
import os
os.chdir('/content/sadie-app')
# Full log is saved to buildozer's log dir; we tail the last lines here.
!buildozer android debug 2>&1 | tail -60
```

### Step 4: Download Your APK! 💜


```
import glob, os
apks = glob.glob('/content/sadie-app/bin/*.apk')
if apks:
    print(f'🎉 APK built successfully: {apks[0]}')
    print(f'📦 Size: {os.path.getsize(apks[0]) / 1024 / 1024:.1f} MB')
    from google.colab import files
    files.download(apks[0])
else:
    print('❌ Build failed — check the output above for errors')
    # Show the tail of the buildozer log to help diagnose
    !cat /content/sadie-app/.buildozer/logs/buildozer.log 2>/dev/null | tail -40
```
