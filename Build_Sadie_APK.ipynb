# 💜 Build Sadie APK
## One-click Android build for your AI proxy agent

**Instructions:**
1. Click **Runtime → Run all** (or press Ctrl+F9)
2. Wait ~15-20 minutes for the build
3. Download your APK from the link at the bottom!

```
!pip install buildozer cython
!sudo apt-get update -qq
!sudo apt-get install -y -qq build-essential git python3-pip autoconf libtool pkg-config zlib1g-dev libncurses5-dev libncursesw5-dev libtinfo5 cmake libffi-dev libssl-dev lld
print('✅ Build tools installed!')
```

### Step 2: Create Sadie's Source Files

```
import os
os.makedirs('/content/sadie-app', exist_ok=True)
os.chdir('/content/sadie-app')
print('📁 Project folder ready')
```

```
# Upload your sadie-agent-setup.zip here
# Option A: Upload manually
from google.colab import files
print('📤 Upload your sadie-agent-setup.zip file:')
uploaded = files.upload()

import zipfile
for fname in uploaded.keys():
    if fname.endswith('.zip'):
        with zipfile.ZipFile(fname, 'r') as z:
            z.extractall('/content/sadie-app/')
        print(f'✅ Extracted {fname}')

# Show what we got
!ls -la /content/sadie-app/
```

### Step 3: Build the APK (~15 min first time)

```
os.chdir('/content/sadie-app')
!buildozer android debug 2>&1 | tail -50
```

### Step 4: Download Your APK! 💜

```
import glob
apks = glob.glob('/content/sadie-app/bin/*.apk')
if apks:
    print(f'🎉 APK built successfully: {apks[0]}')
    print(f'📦 Size: {os.path.getsize(apks[0]) / 1024 / 1024:.1f} MB')
    from google.colab import files
    files.download(apks[0])
else:
    print('❌ Build failed — check the output above for errors')
    # Try to find the error
    !cat /content/sadie-app/.buildozer/logs/buildozer.log 2>/dev/null | tail -30
```
